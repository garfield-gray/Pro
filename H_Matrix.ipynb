```python
import numpy as np

def get_matrix(row, col, sigma):
    # Create a 2D array representing the indices of each element in the matrix

    
    i, j = np.meshgrid(np.arange(1,row+1), np.arange(1,col+1), indexing='ij')

    # Calculate the value of h(i,j) for each element using the indices and sigma
    distance = (i ** 2 + j ** 2) / sigma ** 2
    h =  np.exp(-distance ) / (2 * np.pi * sigma ** 2) 


    # Reverse the order of the rows and columns to match the desired output
    # h = np.fliplr(h)
    # h = np.flipud(h)

    return h
```
